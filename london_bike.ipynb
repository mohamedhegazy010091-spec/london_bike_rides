```python
# Unzip the downloaded file
zip_filename = "london-bike-sharing-dataset.zip"
with zipfile.ZipFile(zip_filename, 'r') as file:
    file.extractall()
```

```python
# Load the dataset using kagglehub
from kagglehub import KaggleDatasetAdapter
# Set the path to the file you'd like to load
file_name ="london_merged.csv"

# Load the latest version
df = kagglehub.dataset_load(
  KaggleDatasetAdapter.PANDAS,
  "hmavrodiev/london-bike-sharing-dataset",
  file_name
)
# Display the first 5 records
print("First 5 records:", df.head())
```

```python
# Create a copy of the dataframe
bike = df.copy()
```

```python
# Rename columns for better understanding
new_col_dict = {
    "timestamp":"time",
    "cnt":"count",
    "t1":"temp_real_c",
    "t2":"temp_feel_c",
    "hum":"humidity_percent",
    "wind_speed":"wind_speed_kmh",
    "weather_code":"weather",
    "is_holiday":"is_holiday",
    "is_weekend":"is_weekend",
    "season":"season",
}
bike.rename(columns=new_col_dict, inplace=True)
bike.head()
```

```python
# Convert humidity from percentage to a fraction
bike.humidity_percent = bike.humidity_percent / 100
bike.head()
```

```python
# Map season to descriptive labels
seson_dict = {
    "0.0":"spring",
    "1.0":"summer",
    "2.0":"autumn",
    "3.0":"winter"
}
# Map weather codes to descriptive labels
weather_dict = {
    "1.0":"clear",
    "2.0":"scattered_clouds",
    "3.0":"broken_clouds",
    "4.0":"cloudy",
    "7.0":"rain",
    "10.0":"rainwith_thunder",
    "26.0":"snowfall"
}
# Convert season column to string type
bike.season = bike.season.astype('str')
# Map season values to descriptive labels
bike.season = bike.season.map(seson_dict)
# Convert weather column to string type
bike.weather = bike.weather.astype('str')
# Map weather values to descriptive labels
bike.weather = bike.weather.map(weather_dict)

```
